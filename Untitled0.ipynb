```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load Data
data = pd.read_csv('tele_sales_mock_data.csv')

# Ensure AgentID is numeric
data['AgentID'] = data['AgentID'].astype(int)

# Display sales per agent
sales_per_agent = data.groupby('AgentID')['SalesAmount'].sum()
print(sales_per_agent)

# Plot sales per agent
plt.figure(figsize=(10,6))
sales_per_agent.plot(kind='bar')
plt.xlabel('AgentID')
plt.ylabel('Sales Amount')
plt.title('Sales per Agent')
plt.show()

# Calculate Conversion Rate
data['ConversionRate'] = (data['SuccessfulCalls'] / data['TotalCalls']) * 100
print(data[['AgentID', 'ConversionRate']])

# Format Conversion Rate as a percentage for display
data['ConversionRateStr'] = data['ConversionRate'].map('{:.2f}%'.format)
print(data[['AgentID', 'ConversionRateStr']])

# Ensure AgentID is treated as a category for discrete x-axis plotting
data['AgentID'] = data['AgentID'].astype(str)  # Convert to string for categorical plotting

# Aggregate Conversion Rate per Agent
conversion_per_agent = data.groupby('AgentID')['ConversionRate'].mean()

# Plot corrected bar chart
plt.figure(figsize=(10,6))
conversion_per_agent.plot(kind='bar', color='skyblue')
plt.title('Conversion Rate by Agent')
plt.xlabel('Agent ID')
plt.ylabel('Conversion Rate (%)')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()

#Seaborn Intro#
sns.histplot(data['SalesAmount'], kde=True)
plt.show()
```
